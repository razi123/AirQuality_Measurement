```python
def retrieveHTML():
    for Year in range(2013, 2019):
        for Month in range(1,13):
            if Month > 9:
                url = "https://en.tutiempo.net/climate/{}-{}/ws-432950.html".format(Month, Year)
            else:
                url = "https://en.tutiempo.net/climate/0{}-{}/ws-432950.html".format(Month, Year)

            html_text = requests.get(url)
            texts_utf = html_text.text.encode('utf=8')

            if not os.path.exists(rootPath + '/AIR_QUALITY_DATA/HTML_FILES/{}'.format(Year)):
                os.makedirs(rootPath + '/AIR_QUALITY_DATA/HTML_FILES/{}'.format(Year))

            with open(rootPath + '/AIR_QUALITY_DATA/HTML_FILES/{}/{}.html'.format(Year,Month),'wb') as output:
                output.write(texts_utf)

            sys.stdout.flush() # writes from internal buffer to file
```

```python
def webScrap(Year, Month):
    fileRead = open(rootPath + '/AIR_QUALITY_DATA/HTML_FILES/{}/{}.html'.format(Year, Month),'rb')
    plainText = fileRead.read()
    
    tempD = []
    finalD = []
    
    soup = BeautifulSoup(plainText,"lxml")
    
    for table in soup.findAll('table', {'class' : 'medias mensuales numspan'}):  # assuming multiple tables
        for tbody in table:
            for tr in tbody:
                text = tr.get_text()
                tempD.append(text)
                
    rows = len(tempD)/15
    
    for times in range(round(rows)):
        newTempD = []
        for i in range(15):
            newTempD.append(tempD[0])
            tempD.pop(0)
        finalD.append(newTempD)
    
    length = len(finalD)
    finalD.pop(length -1)
    finalD.pop(0)
    
    for k in range(len(finalD)):
        finalD[k].pop(14)
        finalD[k].pop(13)
        finalD[k].pop(12)
        finalD[k].pop(11)
        finalD[k].pop(10)
        finalD[k].pop(4)
        finalD[k].pop(0)
    
    return(finalD)       
```

```python
def mergeData(year,cs):
    for a in pd.read_csv(rootPath+ "/Real_data/real_" + str(year) + '.csv', chunksize = cs):
        df = pd.DataFrame(data = a)
        myList = df.values.tolist()
    return myList
```

```python
if not os.path.exists(rootPath + "/Real_Data/"):
    os.makedirs(rootPath + "/Real_Data/")
    
for Year in range(2013,2019):
    finalData = []
    with open(rootPath + "/Real_Data/real_" + str(Year) + '.csv', 'w') as csvFile:
        wr = csv.writer(csvFile, dialect = "excel")
        wr.writerow(
            ['T', 'TM', 'Tm', 'H', 'PP','VV', 'V', 'VM', 'PM2.5'])
        
        for Month in range(1,13):
            temp = webScrap(Year, Month)
            finalData = finalData + temp

        pm = getattr(sys.modules[__name__], 'AQI{}'.format(Year))()

        if len(pm) == 364:
            pm.insert(364,'-')

        for i in range(len(finalData)-1):
            finalData[i].insert(8,pm[i])
            
        with open(rootPath + "/Real_Data/real_" + str(Year) + '.csv', 'a') as csvFile:
            wr = csv.writer(csvFile, dialect= 'excel')

            for row in finalData:
                flag = 0
                for elem in row:
                    if elem == "" or elem == "-":
                        flag=1
                if flag!=1:
                    wr.writerow(row)

data2013 = mergeData(2013,600)
data2014 = mergeData(2014,600)
data2015 = mergeData(2015,600)
data2016 = mergeData(2016,600) 

completeData = data2013 + data2014 + data2015 + data2016 

with open(rootPath + "/Real_data/real_combined.csv", 'w') as csvFile:
    wr = csv.writer(csvFile, dialect = 'excel')
    wr.writerow(['T', 'TM', 'Tm', 'H', 'PP', 'VV', 'V', 'VM', 'PM2.5'])

    wr.writerows(completeData)
```

```python
# Feature Importance
# Extra tree regressor is an Ensembeled leaning algo, has multiple decision trees, donot slice the input just 
# like Random forest but it will consider the whole input
# For regression problem - Arithimitic mean of the predicitions 
# For classification problem - Majority vote 

from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()  # solves the outlier problem
model.fit(X,Y)
```

```python
print(model.feature_importances_)  ## feature importance for PM2.5, sum should be equal to 1
```

```python
feat_imp_series = pd.Series(model.feature_importances_, index = ['T', 'TM', 'Tm', 'H', 'PP', 'VV', 'V', 'VM'])
feat_imp_series.nlargest(7).plot(kind='barh')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
# slope values for independent features , for X
regressor.coef_
```

```python
# Co efficient value of Y when all independent features are 0
regressor.intercept_
```

```python
# Shows the error from the best fir line, value= 1 means the bestfit line passes through all the point
# should be close to 1 or greater than 0.8
print(" Coefficient Regressor determinatin = R^2 <-- on train set: {}".format(regressor.score(X_train, Y_train))) 
```

```python
print(" Coefficient Regressor determinatin = R^2 <-- on train set: {}".format(regressor.score(X_test, Y_test)))
```

```python
# take reuse the existing dataset with different train and test combination, eg: 100/5
from sklearn.model_selection import cross_val_score  
score = cross_val_score(regressor, X, Y, cv= 5)
```

```python
score.mean()  # approximate equal to R^2 
```
